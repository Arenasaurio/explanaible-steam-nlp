```python
from sentence_transformers import SentenceTransformer
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_csv("Datos_Modelo.csv", quoting=3, on_bad_lines='skip')
texts = df["review_text"].astype(str).tolist()
```

```python
#cargar el modelo
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
#truncar textos para no generar errores
texts_truncated = [t[:512] for t in texts]

#generar embeddings
embeddings = model.encode(texts_truncated, show_progress_bar=True, convert_to_numpy=True)
```

```python
#guardar
import numpy as np
np.save("embeddings.npy", embeddings)
```
